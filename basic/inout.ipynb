```python
#random: 파이썬에서 제공하는 패키지

import random

hanguls = list("가나다라마바사아자차카타파하")
f = open("data/info.txt", "w",encoding = "utf-8")

for i in range(500):
    name = random.choice(hanguls) + random.choice(hanguls) # 예_다마
    weight = random.randrange(40,100) #예_90(40-100 랜덤)
    height = random.randrange(140,200) #예_185 (140-200 랜덤)
    f.write("{},{},{}\n".format(name, weight, height))
f.close()
```

```python
#score 읽어와서 평균, 총점 출력
scores = []

f = open("data/score.txt", "r")

line = f.readline()
while line:
    scores.append(int(line))
    print(line, end ="")
    line = f.readline()
f.close()
```

```python
print(scores)
print("합계 : ", sum(scores))
print("평균 : {:.2f}".format(sum(scores) / len(scores))) #	소수 둘째 자리까지 반올림
print("평균 : {:6.3}".format(sum(scores) / len(scores))) #전체 자릿수 6칸, 그 중 유효숫자 3자리

#최신방법(f-string 사용, Python 3.6 이상):
# print(f"평균 : {average:.2f}")
```

```python
#info.txt 읽어와서 BMI 지수 계산 후 화면 출력

#BMI = weight / ((height / 100) ** 2)
# BMI >= 25 : 과체중
# BMI >= 18.5 : 정상체중
# 저체중

#나다 82 180
#이름 : 나다
# 몸무게 : 82
# 키 : 180
# BMI : 33.78
#결과 : 과체중

f = open("data/info.txt", "r",encoding = "utf-8")
for info in f:
    name, weight, height = info.strip().split(",")
    print(name, weight, height)
        
    weight = float(weight)
    height = float(height)

    bmi = weight / ((height / 100) ** 2)

    # 결과 판정
    if bmi >= 25:
        result = "과체중"
    elif bmi >= 18.5:
        result = "정상체중"
    else:
        result = "저체중"

  # 출력
        print(f"이름 : {name}")
        print(f"몸무게 : {weight}")
        print(f"키 : {height}")
        print(f"BMI : {bmi:.2f}")
        print(f"결과 : {result}")
        print("-" * 30)
# f.close()
```
